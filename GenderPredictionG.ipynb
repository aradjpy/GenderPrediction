```python
def findCountry(name):
  c = list(nd.search(name)['first_name']['country'].items())
  max = c[0][1]
  n = 0
  for i in range(1,len(c)):
    if c[i][1] > max:
      n=i
      max = c[i][1]
  return [c[n][0], max]
```

```python
def get_gender(name):
  gender = None
  genderP = None
  try:
    # gender = [i[1] for i in list(nd.search(name)['first_name']['gender'].items())]
    # if gender[0]>gender[1]:
    #   gender = 'Female'
    # else:
    #   gender = 'Male'
    gender = nd.search(name)['first_name']['gender']
    genderP = max(gender.values())
    gender = max(gender, key=gender.get)
  except:
    gender = 'Undefined'
  return [gender,genderP]
```

```python
def fixName(name):
  content = re.split(r'[. ]',name)
  print(content)
  if '' in content:
    content.pop(content.index(''))
    return content[1]
  return content[0]
```

```python
fixName('Mr. Stephen Parsons')
```

```python
df['FirstName'] = df['FullNames'].apply(lambda x: fixName(x))
```
